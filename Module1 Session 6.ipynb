```python
for i in dataset[1:]:
    for j in bad_char:                #inner loop 
        i[6] = i[6].replace(j,"")     #finds and replaces bad char with empty space
    print(i[6])
```

```python
for i in dataset[1:]:
    if i[2] == '':
        i[2] = i[2].replace('','Country not defined')
```

```python

nationality={}                              #Frequency table for nationality of artists

for i in dataset[1:]:
    origin = i[2]
    if origin not in nationality:
        nationality[origin] = 1
    else:
        nationality[origin] +=1

nationality
```

```python
for i in dataset[1:]:
    if i[5].startswith("m"):
        i[5] = i[5].replace("m","M")
    
    if i[5] == '':
        i[5] = i[5].replace('','Gender not defined')
```

```python
gender={}                              #Frequency table for gender of artists

for i in dataset[1:]:

        gene = i[5]
        if gene not in gender:
            gender[gene] = 1
        else:
            gender[gene] +=1
            
```
